```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# URLs de los archivos CSV
url_tienda1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url_tienda2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_2.csv"
url_tienda3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_3.csv"
url_tienda4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar los datos en DataFrames
tienda1 = pd.read_csv(url_tienda1)
tienda2 = pd.read_csv(url_tienda2)
tienda3 = pd.read_csv(url_tienda3)
tienda4 = pd.read_csv(url_tienda4)
```

#1. Análisis de facturación



```python

# Calcular el ingreso total de cada tienda
ingresos = {
    "Tienda 1": tienda1['Precio'].sum(),
    "Tienda 2": tienda2['Precio'].sum(),
    "Tienda 3": tienda3['Precio'].sum(),
    "Tienda 4": tienda4['Precio'].sum()
}

# Crear un DataFrame para la tabla de resultados
df_ingresos = pd.DataFrame(ingresos.items(), columns=['Tienda', 'Ingreso Total'])
print("### Tabla de Ingresos Totales por Tienda")
print(df_ingresos)
print("\n")

# Graficar los ingresos
plt.figure(figsize=(10, 6))
sns.barplot(x='Tienda', y='Ingreso Total', data=df_ingresos)
plt.title('Ingreso Total por Tienda')
plt.ylabel('Ingreso Total ($)')
plt.xlabel('Tienda')
plt.show()
```

# 2. Ventas por categoría

```python

# Función para visualizar las categorías más vendidas
def graficar_categorias(df, nombre_tienda):
    conteo_categorias = df['Categoría'].value_counts().reset_index()
    conteo_categorias.columns = ['Categoría', 'Cantidad Vendida']

    plt.figure(figsize=(12, 7))
    sns.barplot(x='Cantidad Vendida', y='Categoría', data=conteo_categorias)
    plt.title(f'Categorías Más Populares en {nombre_tienda}', fontsize=16)
    plt.xlabel('Cantidad Vendida', fontsize=12)
    plt.ylabel('Categoría', fontsize=12)
    plt.show()

# Aplicar la función a cada tienda
graficar_categorias(tienda1, "Tienda 1")
graficar_categorias(tienda2, "Tienda 2")
graficar_categorias(tienda3, "Tienda 3")
graficar_categorias(tienda4, "Tienda 4")
```

# 3. Calificación promedio de la tienda


```python
# Código corregido para la sección 3. Calificaciones Promedio de los Clientes

# Sustituye '[nombre_correcto_aqui]' por el nombre exacto de la columna de calificaciones.
calificaciones_promedio = {
    "Tienda 1": tienda1['Calificación'].mean(),
    "Tienda 2": tienda2['Calificación'].mean(),
    "Tienda 3": tienda3['Calificación'].mean(),
    "Tienda 4": tienda4['Calificación'].mean()
}

# Crear un DataFrame para la tabla de resultados
df_calificaciones = pd.DataFrame(calificaciones_promedio.items(), columns=['Tienda', 'Calificación Promedio'])
print("### Tabla de Calificaciones Promedio por Tienda")
print(df_calificaciones)
print(
)

# Graficar las calificaciones promedio
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x='Tienda', y='Calificación Promedio', data=df_calificaciones)
plt.title('Calificación Promedio del Cliente por Tienda')
plt.ylim(3.5, 5.0)  # Establecer un límite para mejor visualización de las diferencias
plt.ylabel('Calificación Promedio')
plt.xlabel('Tienda')
plt.show()
```

# 4. Productos más y menos vendidos

```python

# Función para identificar y visualizar los productos más y menos vendidos
def productos_populares(df, nombre_tienda):
    conteo_productos = df['Producto'].value_counts()

    # Top 10 más vendidos
    top_10 = conteo_productos.head(10)

    # Top 10 menos vendidos
    bottom_10 = conteo_productos.tail(10)

    print(f"### Productos Más Vendidos en {nombre_tienda}")
    print(top_10)
    print("\n")

    print(f"### Productos Menos Vendidos en {nombre_tienda}")
    print(bottom_10)
    print("\n")

    # Graficar los resultados
    fig, axes = plt.subplots(1, 2, figsize=(18, 8))

    # Gráfico de más vendidos
    sns.barplot(x=top_10.values, y=top_10.index, ax=axes[0])
    axes[0].set_title(f'Top 10 Productos Más Vendidos en {nombre_tienda}')
    axes[0].set_xlabel('Cantidad Vendida')
    axes[0].set_ylabel('Producto')

    # Gráfico de menos vendidos
    sns.barplot(x=bottom_10.values, y=bottom_10.index, ax=axes[1])
    axes[1].set_title(f'Top 10 Productos Menos Vendidos en {nombre_tienda}')
    axes[1].set_xlabel('Cantidad Vendida')
    axes[1].set_ylabel('Producto')

    plt.tight_layout()
    plt.show()

# Aplicar la función a cada tienda
productos_populares(tienda1, "Tienda 1")
productos_populares(tienda2, "Tienda 2")
productos_populares(tienda3, "Tienda 3")
productos_populares(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
# Calcular el costo de envío promedio
costo_envio_promedio = {
    "Tienda 1": tienda1['Costo de envío'].mean(),
    "Tienda 2": tienda2['Costo de envío'].mean(),
    "Tienda 3": tienda3['Costo de envío'].mean(),
    "Tienda 4": tienda4['Costo de envío'].mean()
}

# Crear un DataFrame para la tabla de resultados
df_envios = pd.DataFrame(costo_envio_promedio.items(), columns=['Tienda', 'Costo de Envío Promedio'])
print("### Tabla de Costo de Envío Promedio por Tienda")
print(df_envios)
print("\n")

# Graficar los costos de envío
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x='Tienda', y='Costo de Envío Promedio', data=df_envios)
plt.title('Costo de Envío Promedio por Tienda')
plt.ylabel('Costo de Envío Promedio ($)')
plt.xlabel('Tienda')
plt.show()
```

```python
# =============================================================================
# CONCLUSIÓN FINAL Y RECOMENDACIÓN
# =============================================================================

print("--- Resumen General de Desempeño por Tienda ---")

# Unir los diferentes análisis en una sola tabla
resumen = df_ingresos.merge(df_calificaciones, on='Tienda').merge(df_envios, on='Tienda')
resumen.set_index('Tienda', inplace=True)

# Presentar la tabla resumen final
display(resumen.style.format({
    'Ingreso Total': '$ {:,.2f}',
    'Calificación Promedio': '{:.2f} ★',
    'Costo de Envío Promedio': '$ {:,.2f}'
}).background_gradient(cmap='RdYlGn', subset=['Ingreso Total', 'Calificación Promedio'], axis=0)
 .background_gradient(cmap='RdYlGn_r', subset=['Costo de Envío Promedio'], axis=0)) # Invertido para costo

# --- Recomendación Estratégica ---
print("\n\n--- RECOMENDACIÓN PARA EL SEÑOR JUAN ---\n")

# Identificar la tienda con el menor ingreso
peor_tienda_ingresos = resumen['Ingreso Total'].idxmin()

# Identificar la tienda con la peor evaluación
peor_tienda_evaluacion = resumen['Calificación Promedio'].idxmin()

# Identificar la tienda con el mayor costo de envío
peor_tienda_envio = resumen['Costo de Envío Promedio'].idxmax()

print("Estimado Señor Juan,")
print("Después de analizar el desempeño de sus cuatro tiendas bajo cinco aspectos clave, hemos llegado a las siguientes conclusiones:\n")
print(f"1.  **Facturación:** La tienda con el ingreso más bajo es **{peor_tienda_ingresos}**.")
print(f"2.  **Satisfacción del Cliente:** La tienda con la peor calificación de los clientes es **{peor_tienda_evaluacion}**.")
print(f"3.  **Costos Logísticos:** La tienda con el costo de envío más alto es **{peor_tienda_envio}**.\n")

# Lógica de recomendación
# Si la misma tienda es la peor en múltiples categorías clave, es la candidata más fuerte.
if peor_tienda_ingresos == peor_tienda_evaluacion:
    print(f"**Recomendación Principal:** La tienda **'{peor_tienda_ingresos}'** presenta el desempeño más bajo tanto en facturación como en la satisfacción de sus clientes. Esta combinación de factores la convierte en la candidata más clara para la venta, ya que requeriría una inversión significativa en marketing y mejoras operativas para ser competitiva.")
else:
    print(f"**Análisis para la Decisión:**")
    print(f"- La tienda **'{peor_tienda_ingresos}'** es la que menos ingresos genera. Si su principal objetivo es deshacerse del activo menos productivo, esta sería la elección.")
    print(f"- La tienda **'{peor_tienda_evaluacion}'** sufre de una baja satisfacción del cliente. Esto representa un riesgo a largo plazo para la reputación y la retención de clientes, incluso si sus ventas actuales son aceptables.")

print("\nNuestra recomendación final es considerar la venta de la **Tienda Iguatemi**. A pesar de no tener la peor calificación, su facturación es significativamente la más baja de todas, lo que limita su capacidad para contribuir a su nueva inversión. Además, su costo de envío es el segundo más alto, lo que reduce aún más su rentabilidad.")
print("\nEsperamos que este análisis detallado le sea de gran utilidad para tomar la mejor decisión estratégica.")
```
